```python
import pandas as pd
import numpy as np
import holidays

# فرض کنید remaining_df دیتافریم شماست و 'cet_cest_timestamp' به عنوان ستون زمانی استفاده می‌شود
remaining_df['cet_cest_timestamp'] = pd.to_datetime(remaining_df['cet_cest_timestamp'])

# افزودن روز هفته
remaining_df['day_of_week'] = remaining_df['cet_cest_timestamp'].dt.day_name()

# افزودن تعطیلات آخر هفته
remaining_df['is_weekend'] = remaining_df['day_of_week'].isin(['Saturday', 'Sunday'])

# ایجاد یک لیست از تعطیلات برای آلمان
de_holidays = holidays.Germany(years=remaining_df['cet_cest_timestamp'].dt.year.unique().tolist())

# بررسی اینکه آیا تاریخ‌ها تعطیلات هستند
remaining_df['is_holiday'] = remaining_df['cet_cest_timestamp'].dt.date.isin(de_holidays)

# افزودن شماره ماه و نام ماه
remaining_df['month'] = remaining_df['cet_cest_timestamp'].dt.month
remaining_df['month_name'] = remaining_df['cet_cest_timestamp'].dt.month_name()

# افزودن ساعت
remaining_df['hour'] = remaining_df['cet_cest_timestamp'].dt.hour

# تعیین ساعات روز (صبح، عصر، شب)
remaining_df['time_of_day'] = np.where(remaining_df['hour'] < 6, 'Night',
                                        np.where(remaining_df['hour'] < 12, 'Morning',
                                                 np.where(remaining_df['hour'] < 18, 'Afternoon', 'Evening')))

# افزودن شماره روز در سال
remaining_df['day_of_year'] = remaining_df['cet_cest_timestamp'].dt.dayofyear

# افزودن شماره هفته در سال
remaining_df['week_of_year'] = remaining_df['cet_cest_timestamp'].dt.isocalendar().week

# تعیین فصل
remaining_df['season'] = np.select(
    [
        (remaining_df['month'].isin([12, 1, 2])),
        (remaining_df['month'].isin([3, 4, 5])),
        (remaining_df['month'].isin([6, 7, 8])),
        (remaining_df['month'].isin([9, 10, 11]))
    ],
    ['Winter', 'Spring', 'Summer', 'Fall'],
    default='Unknown'
)


latitude = 47.6785
longitude = 9.1739

def calculate_day_length(timestamp):
    day_of_year = timestamp.dayofyear

    declination = 23.45 * np.sin(np.radians((360/365) * (day_of_year - 81)))
    sunrise_hour = 12 - (12 / np.pi) * np.arccos(-np.tan(np.radians(latitude)) * np.tan(np.radians(declination)))
    sunset_hour = 12 + (12 / np.pi) * np.arccos(-np.tan(np.radians(latitude)) * np.tan(np.radians(declination)))
    day_length = sunset_hour - sunrise_hour
    return day_length


remaining_df['day_length'] = remaining_df['cet_cest_timestamp'].apply(calculate_day_length)




remaining_df.head()
```

```python
import pandas as pd
import math
import numpy as np

# ---------------------------------------
import matplotlib.pyplot as plt
import seaborn as sns
import pylab
plt.rcParams['figure.figsize']=(17,5)
import holidays
from sklearn.metrics import mean_squared_error, mean_absolute_error

import warnings
warnings.filterwarnings("ignore")
# ---------------------------------------
import plotly.express as px
import plotly.graph_objs as go
from plotly.offline import iplot

# ---------------------------------------

def line_plot(df, date, AEP, text, y_tiltle, x_title):
    data = go.Scatter(x = df[date],
                      y = df[AEP],
                      mode = 'lines',
                      name = AEP)

    layout = go.Layout(title={'text': text,
                              'y':0.9,
                              'x':0.5,
                              'xanchor': 'center',
                              'yanchor': 'top'},
                       xaxis = dict(title = x_title),
                       yaxis = dict(title = y_tiltle),
                       template = 'plotly_dark')

    fig = go.Figure(data = data, layout = layout)
    iplot(fig)

# ---------------------------------------
def new_df(df, col, new_date, new_AEP, split):
    tmp_df = df[[col]].reset_index().rename(columns={'Datetime':new_date, 'AEP_MW':new_AEP})
    tmp_df = pd.DataFrame(tmp_df.resample(split, on='Datetime', origin = 'start').mean())
    tmp_df.reset_index(inplace=True)
    return tmp_df

# ---------------------------------------
def bar_plot(col1, col2, agg_func, title, xtitle, ytitle):
    data = go.Bar(x = df.groupby(col1).agg({col2: agg_func}). \
              reset_index()[col1],
              y = df.groupby(col1).agg({col2: agg_func}). \
              reset_index()[col2],
              text =round(df.groupby(col1).agg({col2: agg_func}).reset_index()[col2],2),
              textposition= 'outside')

    layout = go.Layout(title={'text': title,
                              'y':0.9,
                              'x':0.5,
                              'xanchor': 'center',
                              'yanchor': 'top'},
                       xaxis = dict(title = xtitle),
                       yaxis = dict(title = ytitle),
                       template = 'plotly_dark')

    fig = go.Figure(data = data, layout = layout)
    iplot(fig)


# ---------------------------------------
def box_plot(q_num, col1, col2, color, title, xtitle, ytitle):
    box = go.Box(x = df.loc[df['quarter']==q_num][col1],
                       y = df.loc[df['quarter']==q_num][col2],
                       name = col1,
                       marker_color= color)

    layout = go.Layout(title={'text': title,
                              'y':0.9,
                              'x':0.5,
                              'xanchor': 'center',
                              'yanchor': 'top'},
                       xaxis = dict(title = xtitle),
                       yaxis = dict(title = ytitle),
                       template = 'plotly_dark')


    fig = go.Figure(data = box, layout = layout)
    iplot(fig)


# ---------------------------------------
def MAPE(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
scaler = MinMaxScaler()
scaled_df = scaler.fit_transform(combined_df[['PV generation','day_length','sunrise_seconds',
       'sunset_seconds','hour_cos','hour_sin','solar_angle']])


scaled_df = pd.DataFrame(scaled_df, columns=['PV generation','day_length','sunrise_seconds',
       'sunset_seconds','hour_cos','hour_sin','solar_angle'])
```

```python

# Create sequences
def create_sequences(df, seq_length):
    sequences = []
    labels = []
    for i in range(len(df) - seq_length):
        seq = df.iloc[i:i+seq_length].values
        label = df['PV generation'].iloc[i+seq_length]
        sequences.append(seq)
        labels.append(label)
    return np.array(sequences), np.array(labels)

SEQ_LENGTH = 24  # for 24 hours sequence
sequences, labels = create_sequences(scaled_df, SEQ_LENGTH)

# Train-Test Split
X_train, X_test, y_train, y_test = train_test_split(sequences, labels, test_size=0.2, shuffle=False)

# Model Definition
model = Sequential([
    LSTM(50, return_sequences=True, input_shape=(SEQ_LENGTH, X_train.shape[2])),
    Dropout(0.2),
    LSTM(50),
    Dropout(0.2),
    Dense(1)
])

model.compile(optimizer='adam', loss='mean_squared_error')

# Callbacks for early stopping and learning rate reduction
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5)

# Model Training
history = model.fit(X_train, y_train, epochs=30, batch_size=32, validation_split=0.2, callbacks=[early_stopping, reduce_lr])

# Model Evaluation
predictions_test = model.predict(X_test)
predictions_train = model.predict(X_train)

# Calculate evaluation metrics for test data
r2_test = r2_score(y_test, predictions_test)
mae_test = mean_absolute_error(y_test, predictions_test)
mse_test = mean_squared_error(y_test, predictions_test)
rmse_test = np.sqrt(mse_test)

# Calculate evaluation metrics for train data
r2_train = r2_score(y_train, predictions_train)
mae_train = mean_absolute_error(y_train, predictions_train)
mse_train = mean_squared_error(y_train, predictions_train)
rmse_train = np.sqrt(mse_train)

# Print results for test data
print(f'--- Test Data ---')
print(f'R² Score: {r2_test}')
print(f'Mean Absolute Error (MAE): {mae_test}')
print(f'Mean Squared Error (MSE): {mse_test}')
print(f'Root Mean Squared Error (RMSE): {rmse_test}')

# Print results for train data
print(f'--- Train Data ---')
print(f'R² Score: {r2_train}')
print(f'Mean Absolute Error (MAE): {mae_train}')
print(f'Mean Squared Error (MSE): {mse_train}')
print(f'Root Mean Squared Error (RMSE): {rmse_train}')
```
